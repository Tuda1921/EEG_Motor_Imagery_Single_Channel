```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
import csv

def read_csv_file(file_path):
    data = []
    with open(file_path, 'r') as file:
        csv_reader = csv.reader(file)
        
        for row in csv_reader:
            data.append(row)
    
    return data
```

```python
data = read_csv_file("Data/s1_1.csv")
data = np.array([int(row[0]) for row in data])
plt.plot(data)
plt.show()
```

```python
data = pd.read_csv("Data/s1_1.csv", header=None).transpose()
data = data.values.reshape(-1,512)
data2 = pd.read_csv("Data/s1_2.csv", header=None).transpose()
data2 = data2.values.reshape(-1,512)
data = pd.concat([pd.DataFrame(data), pd.DataFrame(data2)], ignore_index=True)
```

```python
# Tạo DataFrame mới với cột label rỗng
label_data = pd.DataFrame(columns=['label'])

# Đặt số hàng của DataFrame label_data bằng số hàng của DataFrame data
label_data = label_data.reindex(index=range(len(data)))
# Gán nhãn cho từng phần tử trong cột 'label'
for i in range(len(label_data)//2):
    if i % 20 < 10:
        label_data.loc[i, 'label'] = 'right move'
    else:
        label_data.loc[i, 'label'] = 'left move'
for i in range(len(label_data)//2, len(label_data)):
    if i % 20 < 10:
        label_data.loc[i, 'label'] = 'right'
    else:
        label_data.loc[i, 'label'] = 'left'
# In DataFrame label_data
data = pd.concat([data, label_data], axis = 1)
data
```
